```python
import csv
from pathlib import Path

from influxdb_client_3 import Point
from tqdm.notebook import tqdm
```

```python
csv_paths = list(Path("../data/heavy_truck_data").glob("**/*.csv"))

influxdb_lines_protocol = []
for p in tqdm(csv_paths):
    with open(p, mode="r") as f:
        csv_reader = csv.reader(f, delimiter=";")

        for row_id, row in enumerate(csv_reader):
            if row_id == 0:
                continue
            
            binary_data = "".join(["{:08b}" for _ in range(len(row[3:]))]).format(*[int(val) for val in row[3:]])
            
            line_protocol = Point("can_bus").field("id", row[1]).field("dlc", row[2]).field("data", binary_data).time(row[0]).to_line_protocol()
            print(line_protocol)
            # influxdb_lines_protocol.append(
                
            # )
```
